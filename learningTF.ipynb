## Linear regression

```python
import tensorflow as tf,random
with tf.variable_scope("foo",reuse=tf.AUTO_REUSE):
    m_var = tf.get_variable('m',[],initializer=tf.constant_initializer(0.))
    b_var = tf.get_variable('b',[],initializer=tf.constant_initializer(0.))

    input_placeholder = tf.placeholder(tf.float32)
    guess = m_var*input_placeholder + b_var
    output_placeholder = tf.placeholder(tf.float32)

    loss = tf.square(output_placeholder - guess)

    optimizer = tf.train.GradientDescentOptimizer(1e-4)
    train_op = optimizer.minimize(loss)
    sess = tf.Session()
    sess.run(tf.global_variables_initializer())

    m_orig = random.random()
    b_orig = random.random()

    for step in range(100000):
        input_value = random.random()
        output_value = m_orig * input_value + b_orig
        _loss, _ = sess.run([loss,train_op],feed_dict={input_placeholder:input_value,output_placeholder:output_value})
        print(step,_loss)

    print("True parameters: m={:5f} b={:5f}".format(m_orig,b_orig))
    print("Learned parameters: m={:5f} b={:5f}".format(*sess.run([m_var,b_var])))









```

## the tf.Print function

```python
import tensorflow as tf
with tf.variable_scope("bix",reuse=tf.AUTO_REUSE):
    two_node = tf.constant(2)
    three_node = tf.constant(3)
    sum_node = two_node + three_node 
    print_sum_node = tf.Print(sum_node,[two_node,three_node])
    with tf.Session() as sess:
        print(sess.run(print_sum_node))
```

### Graph Inspection

```python
import tensorflow as tf
a = tf.constant(2,name='a')
b = tf.constant(3,name='b')
c = a + b

print("Our tf.Tensor objects:")
print(a,b,c)
a_op = a.op
b_op = b.op
c_op = c.op

print("comp form ")
print(a_op.__repr__())
print(b_op.__repr__())
print(c_op.__repr__())
print("\n")
print(c_op)
print("inspecting consumers")
print(a.consumers(),b.consumers(),c.consumers())

print("input tensors for each op")
print(list(a_op.inputs))

g = tf.get_default_graph()
ops_list = g.get_operations()
print(ops_list)


```
